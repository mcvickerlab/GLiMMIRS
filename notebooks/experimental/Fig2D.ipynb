Plot the distribution of signficant ($q<0.1$) estimated enhancer effects from GLiMMIRS-base for the Gasperini data. 

```r
df <- read.csv('/iblm/netapp/home/karthik/GLiMMIRS/data/gasperini/processed/baseline_models.csv')

df$qval <- p.adjust(df$pvalue.list, 'fdr')
effects.hist <- ggplot(df %>% filter(qval<0.1), aes(x=effect.list)) + geom_histogram(color = 'black', fill = 'darkgray') + xlab(expression(beta[enhancer])) + theme_classic() + theme(text = element_text(size = 18))
effects.hist
```
