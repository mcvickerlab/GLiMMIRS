```r
enhancer.1 <- 'chr11.1735'
enhancer.2 <- 'chr11.1734'
gene <- 'ENSG00000135363'
```

```r
# read in table that connects target sites to guide spacer sequences
target.site.spacers.table <- read.table(
    '/iblm/netapp/data1/external/Gasperini2019/suppl/GSE120861_grna_groups.at_scale.txt',
    sep = '\t'
)
colnames(target.site.spacers.table) <- c('target.site', 'spacer.sequence')
target.site.spacers.table$target.site <- sapply(target.site.spacers.table$target.site, FUN = function(x) {
    if (startsWith(x, 'chr')) {
        return (strsplit(x, '_')[[1]][1])
    }
    else {
        return (x)
    }
})
head(target.site.spacers.table)
```

```r
enhancer.1.spacers <- target.site.spacers.table[target.site.spacers.table$target.site == enhancer.1, ]$spacer.sequence
enhancer.2.spacers <- target.site.spacers.table[target.site.spacers.table$target.site == enhancer.2, ]$spacer.sequence
print(enhancer.1.spacers)
print(enhancer.2.spacers)
```

```r
# read in guide efficiency information
all.guide.efficiencies <- h5read('/iblm/netapp/data1/external/Gasperini2019/processed/gasperini_data.h5', 'guidescan.output')
all.guide.efficiencies$spacer <- substring(all.guide.efficiencies$gRNA, 1, nchar(all.guide.efficiencies$gRNA) - 3)
head(all.guide.efficiencies)
```

```r
enhancer.1.spacers.efficiencies <- all.guide.efficiencies[all.guide.efficiencies$spacer %in% enhancer.1.spacers, c('spacer', 'Cutting.Efficiency')]
enhancer.2.spacers.efficiencies <- all.guide.efficiencies[all.guide.efficiencies$spacer %in% enhancer.2.spacers, c('spacer', 'Cutting.Efficiency')]
print(enhancer.1.spacers.efficiencies)
print(enhancer.2.spacers.efficiencies)
```

```r
enhancer.1.spacers.efficiencies[is.na(enhancer.1.spacers.efficiencies)] <- 0
enhancer.2.spacers.efficiencies[is.na(enhancer.2.spacers.efficiencies)] <- 0
```

```r
# read in cell-guide matrix
cell.guide.matrix <- h5read('/iblm/netapp/data1/external/Gasperini2019/processed/gasperini_data.h5', 'cell.guide.matrix')
guide.spacers <- h5read('/iblm/netapp/data1/external/Gasperini2019/processed/gasperini_data.h5', 'guide.spacers')
print(head(cell.guide.matrix))
print(head(guide.spacers))
```

```r
colnames(cell.guide.matrix) <- guide.spacers
```

```r
enhancer.1.indicator.vector <- rep(1, nrow(cell.guide.matrix))

for (i in 1:nrow(enhancer.1.spacers.efficiencies)) {
    guide.spacer <- enhancer.1.spacers.efficiencies$spacer[i]
    guide.efficiency <- enhancer.1.spacers.efficiencies$Cutting.Efficiency[i]
    guide.indicator.vector <- cell.guide.matrix[, guide.spacer]
    guide.probs <- 1 - (guide.indicator.vector * guide.efficiency)
    enhancer.1.indicator.vector <- enhancer.1.indicator.vector * guide.probs
}

enhancer.1.indicator.vector <- 1 - enhancer.1.indicator.vector
```

```r
enhancer.2.indicator.vector <- rep(1, nrow(cell.guide.matrix))

for (i in 1:nrow(enhancer.2.spacers.efficiencies)) {
    guide.spacer <- enhancer.2.spacers.efficiencies$spacer[i]
    guide.efficiency <- enhancer.2.spacers.efficiencies$Cutting.Efficiency[i]
    guide.indicator.vector <- cell.guide.matrix[, guide.spacer]
    guide.probs <- 1 - (guide.indicator.vector * guide.efficiency)
    enhancer.2.indicator.vector <- enhancer.2.indicator.vector * guide.probs
}

enhancer.2.indicator.vector <- 1 - enhancer.2.indicator.vector
```

```r
# read in counts matrix
gene.counts <- h5read('/iblm/netapp/data1/external/Gasperini2019/processed/gasperini_data.h5', 'gene.counts')
gene.names <- h5read('/iblm/netapp/data1/external/Gasperini2019/processed/gasperini_data.h5', 'gene.names')
rownames(gene.counts) <- gene.names
head(gene.counts)
```

```r
# target.gene.counts <- gene.counts[sample.target.gene, ]
target.gene.counts <- gene.counts[gene, ]
```

```r
interaction.vector <- enhancer.1.indicator.vector * enhancer.2.indicator.vector
interaction.vector
```

```r
enhancer.1.indicator.vector[enhancer.1.indicator.vector > 0] <- 1
enhancer.2.indicator.vector[enhancer.2.indicator.vector > 0] <- 1
interaction.vector[interaction.vector > 0] <- 1
```

```r
enhancer.1.indicator.vector[interaction.vector > 0] <- 0
enhancer.2.indicator.vector[interaction.vector > 0] <- 0
```

```r
non.targeted.vector <- rep(1, nrow(cell.guide.matrix))
non.targeted.vector[enhancer.1.indicator.vector > 0] <- 0
non.targeted.vector[enhancer.2.indicator.vector > 0] <- 0
non.targeted.vector[interaction.vector > 0] <- 0
non.targeted.vector
```

```r
non.targeted.counts <- target.gene.counts[non.targeted.vector > 0]
enhancer.1.counts <- target.gene.counts[enhancer.1.indicator.vector > 0]
enhancer.2.counts <- target.gene.counts[enhancer.2.indicator.vector > 0]
interaction.counts <- target.gene.counts[interaction.vector > 0]
```

```r
max(interaction.counts)
```

```r
non.target.df <- data.frame(non.targeted.counts)
non.target.df$target <- 'non.target'
colnames(non.target.df) <- c('count', 'target')

enhancer.1.df <- data.frame(enhancer.1.counts)
enhancer.1.df$target <- 'enhancer.1'
colnames(enhancer.1.df) <- c('count', 'target')

enhancer.2.df <- data.frame(enhancer.2.counts)
enhancer.2.df$target <- 'enhancer.2'
colnames(enhancer.2.df) <- c('count', 'target')

interaction.df <- data.frame(interaction.counts)
interaction.df$target <- 'interaction'
colnames(interaction.df) <- c('count', 'target')

counts.df <- rbind(non.target.df, enhancer.1.df, enhancer.2.df, interaction.df)
head(counts.df)
```

```r
max(counts.df$count)
```

```r
create.violinplot(
    formula = count ~ target,
    data = counts.df,
    xaxis.cex = 1
)
```
