```python
# import computational packages
import numpy as np
import pandas as pd 

# plotting
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white');
plt.rcParams['figure.figsize'] = (10, 10)

# higher resolution plots
%config InlineBackend.figure_format ='retina'
```

## Cell Cycle Score Distribution Plots

```python
# plot distribution of g2m scores across all cells
ax = plt.gca()
ax.set_title('Distribution of G2M Scores for All Cells')
ax.set_xlabel('G2M Score (from Seurat)')
g2m_scores = pd.read_csv('./../data/g2m_scores.csv', index_col=0)['G2M.Score']
g2m_scores.plot(kind='hist', bins=50);
plt.savefig('g2m_scores.png')
```

```python
# plot distribution of s scores across all cells
ax = plt.gca()
ax.set_title('Distribution of S Scores for All Cells')
ax.set_xlabel('S Score (from Seurat)')
s_scores = pd.read_csv('./../data/s_scores.csv', index_col=0)['S.Score']
s_scores.plot(kind='hist', bins=50);
plt.savefig('s_scores.png')
```
